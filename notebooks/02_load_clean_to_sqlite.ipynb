```python
import pandas as pd
import sqlite3
from pathlib import Path

clean_path = Path("/Users/chowingchan/Desktop/Project/AI-Analytics-Copilot/data/clean")
db_path = Path("/Users/chowingchan/Desktop/Project/AI-Analytics-Copilot/Competitive-Intelligence-Internal-Analytics-System/data/database/thelook_ecommerce.db")

db_path.parent.mkdir(parents=True, exist_ok=True)

conn = sqlite3.connect(db_path)

tables = {
    "distribution_centers": "distribution_centers_clean.csv",
    "events": "events_clean.csv",
    "inventory_items": "inventory_items_clean.csv",
    "order_items": "order_items_clean.csv",
    "orders": "orders_clean.csv",
    "products": "products_clean.csv",
    "users": "users_clean.csv"
}

for table_name, file_name in tables.items():
    df = pd.read_csv(clean_path/file_name)
    df.to_sql(table_name, conn, if_exists="replace", index=False)
    print(f"Loaded {table_name}: {df.shape}")
conn.close()
```
