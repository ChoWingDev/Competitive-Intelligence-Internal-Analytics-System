## Data Cleaning

```python
import pandas as pd
import numpy as np

# convert from object to datetime
def convert_datetime(df, columns):
    for col in columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')
    return df

users = convert_datetime(
    users,
    ['created_at']
)

events = convert_datetime(
    events,
    ['created_at']
)

inventory_items = convert_datetime(
    inventory_items,
    ['created_at', 'sold_at']
)

order_items = convert_datetime(
    order_items,
    ['created_at', 'shipped_at', 'delivered_at', 'returned_at']
)

orders = convert_datetime(
    orders,
    ['created_at', 'shipped_at', 'delivered_at', 'returned_at']
)
```

```python
# check data types
for name, df in {
    'events': events,
    'inventory_items': inventory_items,
    'order_items': order_items,
    'orders': orders,
    'users': users
}.items():
    print(f"n/{name}")
    display(df.dtypes)
```

## Missing Value Analysis

```python
def missing_summary(df):
    missing_count = df.isnull().sum()
    missing_pct = (missing_count/len(df)) * 100
    
    summary = pd.DataFrame({
        'missing_count': missing_count,
        'missing_pct' : missing_pct
    })
    
    summary = summary [
        summary['missing_count']>0
    ].sort_values(
        by='missing_pct',
        ascending=False
    )
    return summary

for name, df in {
    'distribution_centers': distribution_centers,
    'events': events,
    'inventory_items': inventory_items,
    'order_items': order_items,
    'orders': orders,
    'products': products,
    'users': users
}.items():
    print(f"n/{name}")
    display(missing_summary(df))
```

## Observation

No negative / 0 pricing values where detected in both order_items and products tables, indicating generally valid pricing records without obvious refund or dirty data issues.

No cases were found where delivered at occurred before shipped_at, and no cases where returned_at occurred before delivered_at, suggesting the delivery lifecycle is mostly logically consistent.

Negative delivery_days values ranging from -1 to -4 days indicate potential timestamp inconsistencies, ETL synchronization issues, or data generation anomalies within the dataset.

Since delivered_at should logically occur after order creation, these records are considered invalid for delivery performance analysis and may distort fulfillment KPI calculations.

Recommended Handling:
For downstream analytics and KPI reporting, records with delivery_days < 0 should either:
1. Be excluded from fulfillment-related analysis, or
2. Be corrected only if reliable business rules or source system validation are available.

Delivery duration analysis showed an average delivery time of approximately 2.5 days, with most deliveries completed within 1–4 days, indicating relatively efficient fulfillment operations for valid records.

Customer age distribution appears reasonable, ranging from 12 to 70 years old, with no obvious abnormal age values detected.

No duplicate SKU values were detected in the products table, indicating strong product key uniqueness.

Approximately 16,083 duplicated email values were identified in the users table. This may represent shared/family accounts, synthetic demo data generation, or non-unique customer identities, and should be considered during customer-level analysis.

# Schema and Grain Documentation

| Table                  | Grain                               | Primary Key | Main Foreign Keys                                                                                                                | Business Meaning                                                 |
| ---------------------- | ----------------------------------- | ----------- | -------------------------------------------------------------------------------------------------------------------------------- | ---------------------------------------------------------------- |
| `users`                | One row per registered user         | `id`        | N/A                                                                                                                              | Customer profile and acquisition source                          |
| `orders`               | One row per order                   | `order_id`  | `user_id` → `users.id`                                                                                                           | Order-level status, timestamps, and item count                   |
| `order_items`          | One row per item within an order    | `id`        | `order_id` → `orders.order_id`; `user_id` → `users.id`; `product_id` → `products.id`; `inventory_item_id` → `inventory_items.id` | Item-level sales, shipping, delivery, return, and sale price     |
| `products`             | One row per product SKU             | `id`        | `distribution_center_id` → `distribution_centers.id`                                                                             | Product catalog, category, brand, cost, and retail price         |
| `inventory_items`      | One row per physical inventory unit | `id`        | `product_id` → `products.id`; `product_distribution_center_id` → `distribution_centers.id`                                       | Inventory lifecycle, product copy, cost, and sold timestamp      |
| `distribution_centers` | One row per distribution center     | `id`        | N/A                                                                                                                              | Warehouse / fulfillment center location                          |
| `events`               | One row per user/session web event  | `id`        | `user_id` → `users.id` nullable                                                                                                  | Clickstream event data for browsing, cart, and purchase behavior |


## Key Grain Notes

- `orders` is the order-level table. Each row represents one customer order.
- `order_items` is more granular than `orders`. One order can contain multiple order items.
- `products` is the product catalog table. It should be joined to `order_items` using `product_id`.
- `inventory_items` represents physical inventory units and should be joined using `inventory_item_id` when analyzing inventory lifecycle.
- `events` is clickstream-level data and may include anonymous sessions, so `user_id` can be missing.

## Join Warning

When calculating revenue, use `order_items` as the base table because `sale_price` is stored at item level.

Avoid joining `orders` to `order_items` and then counting rows as orders, because this will duplicate orders with multiple items.

Correct logic:
- Revenue = sum of `order_items.sale_price`
- Order count = count distinct `orders.order_id`
- Item count = count rows in `order_items`
